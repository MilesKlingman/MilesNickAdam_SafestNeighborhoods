**CODE**

```python
import pandas as pd
import matplotlib as plt
import numpy as np
import csv
```

```python
Fire_Incidents = pd.read_csv("Fire_Incidents.csv", index_col = "_id")
Fire_Incidents.head()
```

```python
# Plotting the neighborhoods with above 80 incidents
incidents = {}
querey_mask = Fire_Incidents['neighborhood'].value_counts() > 80
incidents = Fire_Incidents['neighborhood']
incidents = incidents.value_counts()
incidents = incidents[querey_mask]
incidents.plot.bar()
```

```python
# Plotting the neighbirhoods with below 80 incidents
incidents = {}
querey_mask = Fire_Incidents['neighborhood'].value_counts() < 80
incidents = Fire_Incidents['neighborhood']
incidents = incidents.value_counts()
incidents = incidents[querey_mask]
incidents.plot.bar()
```

```python
# Plotting the 10 neighborhoods with the least amount of fire incidents
incidents = {}
querey_mask = Fire_Incidents['neighborhood'].value_counts() < 20
incidents = Fire_Incidents['neighborhood']
incidents = incidents.value_counts()
incidents = incidents[querey_mask]
incidents.plot.bar()
```
